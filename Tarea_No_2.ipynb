# Glosario
**Gota (Blob):** Es un archivo con muchos bytes
**Árbol (Tree):** Es un directorio que contiene archivos y subdirectorios con nombre
**Confirmaciones (Commit):** Tiene padres, metadatos y el árbol de nivel superior
**Instantánea (Snapshot):** El árbol de nivel superior
**Referencias:** Nombres simples que se les dan a las Hash para acceder a las confirmaciones
**Repositorio:** Son los Datos, Objetos y Referencias de Git
**Git-DAG:** Visualizador de Git más amigable 
**GUI:** Interfaces gráficos para Git si no se quiere usar comandos en línea
**Shell:** Líneas de comando de Git 
**Editor:** integraciones prácticas con muchas características (Vim).
**GitHub:** Git no es GitHub. GitHub contribuye con código par otros proyectos (solicitudes de extracción)

# Redireccionamiento de contenidos 
En Git todos los objetos se direccionan por contenido a través de un algoritmo de criptográfico de 160 bites llamado SHA-1 y es un hash. Blob, Tree y Commit son objetos y aunque no los contienen, acceden a ellos a través del hash.

# Referencias 
Un hash es algo difícil de recordar, entonces Git utiliza referencias (Nombres fáciles de recordar) para las confirmaciones. Los objetos no pueden ser alterados mientras que las referencias sí y apuntan a las confirmaciones. 

Saber donde estamos actualmente en el historial se puede saber a través de HEAD y es una referencia

## Comandos en línea de Git

*--Lo esencial--*

git help <command>: obtenga ayuda para un comando git

git init: crea un nuevo repositorio de git, con datos almacenados en el .gitdirectorio

git status: te dice lo que está pasando

git add <filename>: agrega archivos al área de preparación

git commit: crea una nueva confirmación

git log: muestra un registro de historial aplanado

git log --all --graph --decorate: visualiza la historia como un DAG

git diff <filename>: muestra los cambios realizados en relación con el área de preparación

git diff <revision> <filename>: muestra diferencias en un archivo entre instantáneas

git checkout <revision>: actualiza HEAD y la rama actual

    
*--Ramificación y fusión--*
    
git branch: muestra ramas

git branch <name>: crea una rama

git checkout -b <name>: crea una rama y cambia a ella igual que git branch <name>; git checkout <name>

git merge <revision>: se fusiona con la rama actual

git mergetool: use una herramienta elegante para ayudar a resolver conflictos de fusión

git rebase: reensamblar el conjunto de parches en una nueva base

*--Mandos a distancia--*
    
git remote: lista de mandos a distancia

git remote add <name> <url>: agregar un control remoto

git push <remote> <local branch>:<remote branch>: envía objetos a remoto y actualiza la referencia remota

git branch --set-upstream-to=<remote>/<remote branch>: configurar la correspondencia entre la sucursal local y remota

git fetch: recuperar objetos / referencias de un control remoto

git pull: igual que git fetch; git merge

git clone: descargar el repositorio de forma remota

*--Deshacer--*
    
git commit --amend: editar el contenido / mensaje de una confirmación

git reset HEAD <file>: desmontar un archivo

git checkout -- <file>: descartar los cambios

*--Git avanzado--*
    
git config: Git es altamente personalizable

git clone --depth=1: clon superficial, sin historial de versiones completo

git add -p: puesta en escena interactiva

git rebase -i: rebase interactivo

git blame: muestra quién editó por última vez qué línea

git stash: eliminar temporalmente las modificaciones al directorio de trabajo

git bisect: historial de búsqueda binaria (por ejemplo, para regresiones)

.gitignore: especificar archivos no rastreados intencionalmente para ignorar.

# Conclusiones de lo que aprendí: 
1. Git no es Github 
2. Git es un gestor de versiones de código fuente y las entidades que los desarrollan
3. Tiene una estructura jerárquica y permite tener gotas, árboles, confirmaciones de las instantáneas
4. Maneja referencias para acceder a los objetos 
5. Tiene inteligencia para administrar cambios paralelos al mismo objeto por diferentes programadores
6. Permite áreas de ensayo para asegurar la última versión antes de hacer una instantánea
